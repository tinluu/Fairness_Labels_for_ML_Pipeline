```python
import pandas as pd
import numpy as np
import seaborn as sns
pd.options.display.max_rows=999
```

```python
data = pd.read_csv('../data/adult.csv')
```

```python
sensi_atts = ["workclass", "race", "sex"]
target_name = "income-per-year"
groupby_cols = sensi_atts + [target_name]
```

```python
df = data[groupby_cols].groupby(sensi_atts)
groups = list(df.groups.keys())
```

```python
labels = {}
for group in groups:
    dist = df.get_group(group)[target_name].value_counts()/df.get_group(group)[target_name].count()
    labels[group] = dict(round(dist, 3))
```

```python
static_statistics = pd.DataFrame(labels.values())
attr = pd.DataFrame(labels.keys())
attr.columns = sensi_atts
static_table = pd.concat([attr, static_statistics], sort = False, axis = 1).fillna(0)
static_table
```

```python
static_table.groupby(sensi_atts).first()
```

### Say we are interested in the Federal-gov work class, and want to see the distribution of race and gender with regards to the target, here's the visualization:

```python
group_interest = static_table[static_table["workclass"]=="Federal-gov"]
group_interest
```

```python
sns.barplot(x = "race", y = "<=50K", hue = "sex", data = group_interest)
```
